```python
import pandas as pd
import plotly.express as px

CSV_FILE = 'casesstate.csv'
df = pd.read_csv(CSV_FILE)
df['date'] = pd.to_datetime(df['date'])
unique_years = df['date'].dt.year.unique()
unique_months = df['date'].dt.month.unique()

def validate_years_and_months(year, month):
    if (
        year in unique_years and 2020 <= year <= 2023 and
        month in unique_months and 1 <= month <= 12
    ):
        return True
    else:
        return False

while True:
    year_input = input("Step 1: Please enter a year (2020-2023) or press Enter to skip: ")
    if not year_input:
        df_filtered = df
        break
    
    month_input = input("Step 2: Please enter a month (1-12) or press Enter to skip: ")
    if not month_input:
        df_filtered = df
        break
    
    try:
        year = int(year_input)
        month = int(month_input)
        if validate_years_and_months(year, month):
            df_filtered = df[
                (df['date'].dt.year == year) &
                (df['date'].dt.month == month)
            ]
            total_cases = df_filtered['cases_new'].sum()  # Calculate total cases
            print(f"Total cases for {year}-{month} is {total_cases} cases")
            break
        else:
            print("Invalid year or month. Please try again.")
    except ValueError:
        print("Invalid input. Please enter a valid year and month.")
        
if 'df_filtered' in locals():
    fig = px.line(
        df_filtered,
        x='date',
        y='cases_new',
        color='state',
        title='Trend of New COVID-19 Cases in Different States by Year-Month'
    )
    fig.update_layout(xaxis_title='Date', yaxis_title='New Cases')
    fig.show()
```
